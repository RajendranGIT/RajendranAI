STACK

```python
# stack is a collection of items that follows the Last In First Out (LIFO) principle    
# stack LIFO - Using for Browser history, undo functionality in text editors, call stack in programming languages

browser_history = []
browser_history.append('google.com')
browser_history.append('facebook.com')
browser_history.append('twitter.com')
print(browser_history)
```

```python
browser_history.append('python.org')  # Remove the last visited site
print(browser_history)
```

QUEUE - FIFO

```python
# Queue is a collection of items that follows the First In First Out (FIFO) principle
# queue FIFO - Using for print jobs, task scheduling, customer service systems

from collections import deque
ticket_queue = deque()
ticket_queue.append('Customer 1') # Add a customer to the queue Enqueue
ticket_queue.append('Customer 2')
ticket_queue.append('Customer 3')
print(ticket_queue)
```

```python
servered_customer = ticket_queue.popleft()  # Serve the first customer in the queue Dequeue
print(f'Served: {servered_customer}')
```

Named Tuple

```python
# Named tuple is a subclass of tuple that allows you to access elements by name instead of index

from collections import namedtuple
transaction = namedtuple('Transaction', ['id', 'amount', 'type'])
t1 = transaction(id=1, amount=100, type='debit')
print(t1)
```
