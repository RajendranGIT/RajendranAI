```python
import pandas as pd
import numpy as np
 
# Table 1: Employee info
employees = pd.DataFrame({
    'emp_id'  : [101, 102, 103, 104, 105],
    'name'    : ['Alice','Bob','Carol','Dave','Eve'],
    'dept_id' : [10, 20, 10, 30, 20],
})
 
# Table 2: Department info
departments = pd.DataFrame({
    'dept_id'  : [10, 30, 40],
    'dept_name': ['Engineering','HR','Finance'],
    'location' : ['NYC','Chicago','Boston'],
})
 
inner = pd.merge(employees, departments, on='dept_id', how='inner')
print("Inner Join:\n", inner)
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
image = Image.open('dog.jpg')
image_array = np.array(image)
plt.imshow(image_array)
 
```

```python
print(image_array.shape)
```

```python
print(image_array)
```

```python
cropped = image_array[200:800, 150:450]
plt.imshow(cropped)
```

```python
flipped = np.flip(image_array, axis=1)
plt.imshow(flipped)
```

```python
flipped1 = np.flipud(image_array)
plt.imshow(flipped1)
```

```python
rotated = np.rot90(image_array)
plt.imshow(rotated)
```

```python
brightness = image_array + 50
brightness = np.clip(brightness, 0, 255)
plt.imshow(brightness)
```
