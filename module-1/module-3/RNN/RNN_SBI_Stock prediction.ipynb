# **SBI Stock Price Prediction using Neural Network during indian festival sessions**
**Recurrent Neural network provides best outcome this use case compare than Neural Network with Dense layer. Better to go with RNN in this use case**
Explanation:
1.	It imports necessary libraries such as Pandas, NumPy, Keras, and Matplotlib.
2.	A list of festivals with their respective dates is provided.
3.	The get_date_range function calculates a 10-day window before and after each festival date.
4.	For each festival, a date range (start and end) is generated.
5.	The fetch_and_train_model function fetches stock data for the specified range and normalizes it and below technique are used

**The model consists of two SimpleRNN layers, each with 50 neurons and ReLU activation, designed to capture temporal dependencies in stock price data.
The first RNN layer has the argument return_sequences=True to pass the output to the next RNN layer.
The second RNN layer processes the output from the first RNN layer, reducing the dimensionality of the data.
The output layer is a Dense layer with a single neuron, predicting the stock price (a continuous value).
The model is trained using 50 epochs with a batch size of 20, aiming to optimize the weights and minimize the loss using the Adam optimizer.**

6.	The stock data is preprocessed, where features (X) and target labels (y) are prepared.
7.	The data is split into training and testing sets, and a neural network model is built using Keras.
8.	The model is compiled and trained with the stock data.



# Gathering India Festival dates of 2024

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN
import matplotlib.pyplot as plt
import yfinance as yf

# List of festivals with their dates
# Festivals that have already occurred in 2026 (as of April 19)
festivals = [
    {"name": "Makar Sankranti",           "date": "2026-01-14"},
    {"name": "Vasant Panchami",           "date": "2026-02-02"},
    {"name": "Maha Shivratri",            "date": "2026-02-15"},
    {"name": "Holika Dahan",              "date": "2026-03-03"},
    {"name": "Holi",                      "date": "2026-03-04"},
    {"name": "Eid al-Fitr",               "date": "2026-03-30"},
    {"name": "Good Friday",               "date": "2026-04-03"},
    {"name": "Easter",                    "date": "2026-04-05"},
    {"name": "Ram Navami",                "date": "2026-03-27"},
    {"name": "Baisakhi",                  "date": "2026-04-14"},
    {"name": "Vishu",                     "date": "2026-04-14"},
]


# Calculate date ranges for each festival
def get_date_range(festival_date):
    festival_date = datetime.strptime(festival_date, "%Y-%m-%d")
    start_date = festival_date - timedelta(days=10)
    end_date = festival_date + timedelta(days=10)
    return start_date, end_date

# Prepare date ranges
date_ranges = [
    {
        "festival": festival["name"],
        "festival_date": festival["date"],
        "start_date": get_date_range(festival["date"])[0],
        "end_date": get_date_range(festival["date"])[1],
    }
    for festival in festivals
]

# Function to fetch stock data and train the model
def fetch_and_train_model(date_range):
    print(
        f"Processing Festival: {date_range['festival']}, "
        f"Start Date: {date_range['start_date'].date()}, End Date: {date_range['end_date'].date()}"
    )

    try:
        # Fetch stock data
        data = yf.download(
            'SBIN.NS', start=date_range['start_date'].date(), end=date_range['end_date'].date()
        )
        data = data[['Close']]

        # Normalize the data
        scaler = MinMaxScaler(feature_range=(0, 1))
        scaled_data = scaler.fit_transform(data)

        # Prepare training data
        X = scaled_data[:-1]  # All except the last value as input
        y = scaled_data[1:]  # All except the first value as target
        X = X.reshape(X.shape[0], 1, 1)  # Reshape to (samples, timesteps, features)

        # Split into train and test sets
        train_size = int(len(X) * 0.8)
        X_train, X_test = X[:train_size], X[train_size:]
        y_train, y_test = y[:train_size], y[train_size:]

        # Build the RNN model using SimpleRNN
        model = Sequential([
            SimpleRNN(50, activation='relu', return_sequences=True, input_shape=(X_train.shape[1], 1)),
            SimpleRNN(50, activation='relu'),  # Second SimpleRNN layer
            Dense(1)  # Output layer
        ])

        # Compile and train the model
        model.compile(loss='mean_squared_error', optimizer='adam')
        model.fit(X_train, y_train, epochs=50, batch_size=20, verbose=0)

        # Predict
        predicted_prices = model.predict(X_test)
        predicted_prices = scaler.inverse_transform(predicted_prices.reshape(-1, 1))
        actual_prices = scaler.inverse_transform(y_test.reshape(-1, 1))

        return data.index[-len(y_test):], actual_prices, predicted_prices
    except Exception as e:
        print(f"Error processing {date_range['festival']}: {e}")
        return [], [], []

# Plot all festivals in a single chart with subplots
def plot_all_festivals(date_ranges):
    num_festivals = len(date_ranges)
    fig, axes = plt.subplots(num_festivals, 1, figsize=(15, 5 * num_festivals), sharex=True)

    for i, date_range in enumerate(date_ranges):
        dates, actual_prices, predicted_prices = fetch_and_train_model(date_range)
        if len(dates) > 0:
            ax = axes[i] if num_festivals > 1 else axes
            ax.plot(dates, actual_prices, label="Actual Prices", color='blue')
            ax.plot(dates, predicted_prices, linestyle='--', label="Predicted Prices", color='orange')

            # Calculate and plot a horizontal line at the mean of actual prices
            mean_price = actual_prices.mean()
            ax.axhline(mean_price, color='green', linestyle='--', label=f"Mean Price: {mean_price:.2f}")

            ax.set_title(f"{date_range['festival']} ({date_range['festival_date']})")
            ax.set_xlabel("Date")
            ax.set_ylabel("Price")
            ax.legend()
            ax.grid(True)

    plt.tight_layout()
    plt.show()


# Plot the chart
plot_all_festivals(date_ranges)
```
